```python
def parse_maybe_int(i):
    if i=='':
        return None
    else:
        
        return int(float(i))
```

```python
for data in passengers:
     data['PassengerId']=parse_maybe_int(data['PassengerId'])
     data['Pclass']=parse_maybe_int(data['Pclass'])
     data['Age']=parse_maybe_int(data['Age'])
     data['SibSp']=parse_maybe_int(data['SibSp'])
     data['Parch']=parse_maybe_int(data['Parch'])
     data['Fare']=parse_maybe_int(data['Fare'])
     if(data['Survived']=='1' ):
       data['Survived']=True
     else:
       data['Survived']=False
       
    
```

```python
def survivor_number(passengers):
    total=0
    survive_list=[]
    for data in passengers:
        total+=data['Survived']
    survive_list.append(len(passengers))  
    survive_list.append(total)
    return(survive_list)    
```

```python

from collections import defaultdict

def gender_classifier(datas):
    passengers_s=defaultdict(list)
    for data in datas:
        if data['Sex']=='male':
            passengers_s['male'].append(data)
        else:
            passengers_s['female'].append(data) 
    return passengers_s
             
```

```python
import numpy as np
import matplotlib.pyplot as plt
def graph_plotter(n,data1,data2,ylabel,title,xticklabels,legend):
#     N = 2
#     men_means = [survive_men[1], survive_women[1]]
    # men_std = (2, 3)

    ind = np.arange(n)  # the x locations for the groups
    width = 0.35       # the width of the bars

    fig, ax = plt.subplots(figsize=(20, 10))
    rects1 = ax.bar(ind, data1, width, color='r')

#     women_means = [survive_men[0], survive_women[0]]
    # women_std = (3, 5)
    rects2 = ax.bar(ind + width, data2, width, color='y')

    # add some text for labels, title and axes ticks
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    ax.set_xticks(ind + width / 2)
    ax.set_xticklabels(xticklabels)

    ax.legend((rects1[0], rects2[0]), legend)


    def autolabel(rects):
        """
        Attach a text label above each bar displaying its height
        """
        for rect in rects:
            height = rect.get_height()
            ax.text(rect.get_x() + rect.get_width()/2., 1.0*height,
                    '%d' % int(height),
                    ha='center', va='bottom')

    autolabel(rects1)
    autolabel(rects2)

    plt.show()
```

```python
from collections import defaultdict
passenger_age=defaultdict(list)

for data in passengers:
    if type(data['Age']) is int:
        if data['Age']>59:
            passenger_age['old'].append(data)
        elif data['Age']>24:   
             passenger_age['adult'].append(data)
        elif data['Age']>14:   
             passenger_age['young'].append(data)        
        else:
             passenger_age['child'].append(data) 
             
     
```

```python
import pandas as pd



def correlation(x, y):
    std_x=(x-x.mean())/x.std(ddof=0)
    std_y=(y-y.mean())/y.std(ddof=0)
    '''
    Fill in this function to compute the correlation between the two
    input variables. Each input is either a NumPy array or a Pandas
    Series.
    
    correlation = average of (x in standard units) times (y in standard units)
    
    Remember to pass the argument "ddof=0" to the Pandas std() function!
    '''
    return (std_x*std_y).mean()

survived = passengers_df['Survived']
pclass = passengers_df['Pclass']
age=passengers_df['Age']

# rain = subway_df['meanprecipi']
# temp = subway_df['meantempi']

print(correlation(survived, pclass))
print(correlation(survived, age))
# print(correlation(entries, temp))
# print(correlation(rain, temp))

# print(correlation(entries, cum_entries))
```
